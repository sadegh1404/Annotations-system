```python
import requests
import bs4 
import pandas as pd
```

```python
baseUrl = 'https://www.farsnews.com'
category = ['/economy','/world','/politics','/scientific-academic','/culture','/sports','/province']
```

```python
def findPageNumbers(cat):
    """Find number of page of any news category in farsnews.com"""
    
    page = requests.get(baseUrl+cat)
    
    soup = bs4.BeautifulSoup(page.text,'lxml')
    
    n = soup.find('input',{'id':'NewsCount'})
    
    return int(n['value'])
```

```python

def findLinks(cat,count):
    """get the category and find all news links"""
    
    links = []
    
    for pageNumber in range(1,count+1):
    
        mainUrl = baseUrl + cat +'?p='+ str(pageNumber)
        
        c = requests.get(mainUrl)
        content = bs4.BeautifulSoup(c.text,'lxml')
        
        news = content.findAll('li','align-items-start')
        
        for i in news:
            link = i.find('a')
            links.append(baseUrl+link['href'])
    
    return links
```

```python
def getNews(links,cat):
    data = []
    
    for link in links:
        
        print(link)
        news = requests.get(link)
        news = bs4.BeautifulSoup(news.text,'lxml')
        mainPart = news.find('div','nt-frame')
        
        try:
            title = mainPart.find('span','title').text
        except:
            title = 'Nan'
        try:
            abstract = mainPart.find('p','lead').text

        except:
            abstract = 'NaN'
        try:
            imgsrc = mainPart.find('img','w-100')['src']
        except:
            imgsrc = "Nan"
    

        mText = ''
        try:
            for p in mainPart.findAll('p','rtejustify'):
                mText += p.text + '\n'
        except:
            mText = "Nan"
        data.append((title.strip(),abstract.strip(),imgsrc.strip(),mText.strip()))
        
    d = pd.DataFrame(data,columns=['title','abstract','imgsrc','mText'])
    d.to_csv('/home/sadegh/Project/'+cat+'.csv',index=False,encoding='utf-8')
```

```python
for cat in category[4:]:
    print(cat)
    count = findPageNumbers(cat)
    links = findLinks(cat,count)
    getNews(links,cat)
    
```
